```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.feature import greycomatrix, greycoprops

skin_dir = "/content/drive/MyDrive/Dataset/Train/Psoriasis"
skin_list=["/content/drive/MyDrive/Dataset/Train/Psoriasis"]
skin_lower = np.array([0, 20, 70], dtype=np.uint8)
skin_upper = np.array([20, 255, 255], dtype=np.uint8)
glcm_props = ['contrast', 'homogeneity', 'energy']
hue_values = []
texture_values = []
shape_values = []
erythema_values = []
IMAGE_PATH = []
LABEL = []

for index in range(len(skin_list)):
  for image in os.listdir(skin_list[index]):
    IMAGE_PATH.append(os.path.join(skin_list[index], image))

for i in range(3979):
  LABEL.append("Psoriasis")


database = pd.DataFrame(IMAGE_PATH, columns=['path'])
database['label'] = LABEL


for filename in os.listdir(skin_dir):
  if filename.endswith(".jpg") or filename.endswith(".png"):
# Load skin image
    img = cv2.imread(os.path.join(skin_dir, filename))
    # Convert image to HSV color space
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Create a mask to filter out non-skin pixels
    skin_mask = cv2.inRange(hsv_img, skin_lower, skin_upper)

    # Calculate average hue value of skin pixels
    skin_hues = hsv_img[:, :, 0][skin_mask > 0]
    avg_hue = np.mean(skin_hues)

    # Append average hue value to list of color features
    hue_values.append(avg_hue)

    # Convert image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Calculate GLCM matrix and properties
    glcm = greycomatrix(gray_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    glcm_props_vals = [greycoprops(glcm, prop)[0, 0] for prop in glcm_props]

    # Append GLCM properties to list of texture features
    texture_values.append(glcm_props_vals)

    # Calculate image shape features
    moments = cv2.moments(gray_img)
    hu_moments = cv2.HuMoments(moments).flatten()

    # Append Hu moments to list of shape features
    shape_values.append(hu_moments)

    # Calculate erythema feature
    b, g, r = cv2.split(img)
    mean_intensity = np.mean(r)
    erythema_val = np.sum((r-mean_intensity)**2)/(img.shape[0]*img.shape[1])
    erythema_values.append(erythema_val)

color_df = pd.DataFrame({"hue_values": hue_values})
texture_df = pd.DataFrame(texture_values, columns=glcm_props)
shape_df = pd.DataFrame(shape_values, columns=["hu_{}".format(i) for i in range(7)])
erythema_df = pd.DataFrame({"erythema_values": erythema_values})

df = pd.concat([color_df, texture_df, shape_df,erythema_df, database], axis=1)

# Save DataFrame as CSV file
df.to_csv("skin_color_texture_shape_features.csv", index=False)
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from skimage.feature import greycomatrix, greycoprops
import cv2
from sklearn.ensemble import RandomForestClassifier
from skimage import io, color, feature, morphology
#Load the CSV file containing the features and labels

df = pd.read_csv('/content/new_file.csv')
#Extract the features and labels from the CSV file

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
#Convert features to unsigned integer type

X = X.astype('uint8')
#Train a random forest classifier on the data

rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X, y)
#Define a function to extract features from a new image

def extract_features(image_path):
    skin_lower = np.array([0, 20, 70], dtype=np.uint8)
    skin_upper = np.array([20, 255, 255], dtype=np.uint8)
    glcm_props = ['contrast', 'homogeneity', 'energy']
    hue_values = []
    texture_values = []
    shape_values = []
    erythema_values = []
    img = cv2.imread(image_path)
    # Convert image to HSV color space
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Create a mask to filter out non-skin pixels
    skin_mask = cv2.inRange(hsv_img, skin_lower, skin_upper)

    # Calculate average hue value of skin pixels
    skin_hues = hsv_img[:, :, 0][skin_mask > 0]
    avg_hue = np.mean(skin_hues)

    # Append average hue value to list of color features
    hue_values.append(avg_hue)

    # Convert image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Calculate GLCM matrix and properties
    glcm = greycomatrix(gray_img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True)
    glcm_props_vals = [greycoprops(glcm, prop)[0, 0] for prop in glcm_props]

    # Append GLCM properties to list of texture features
    texture_values += glcm_props_vals

    # Calculate image shape features
    moments = cv2.moments(gray_img)
    hu_moments = cv2.HuMoments(moments).flatten()

    # Append Hu moments to list of shape features
    shape_values += hu_moments.tolist()

    # Calculate erythema feature
    b, g, r = cv2.split(img)
    mean_intensity = np.mean(r)
    erythema_val = np.sum((r-mean_intensity)**2)/(img.shape[0]*img.shape[1])
    erythema_values.append(erythema_val)

    return np.array(hue_values + texture_values + shape_values + erythema_values)
def predict_disease(image_path):
    features = extract_features(image_path)
    #print(features)
    prediction = rfc.predict([features])[0]
    return prediction
```
